## Ciclos de Busca e Execução

Um programa de computador é uma lista de instruções, e essas instruções dependem da arquitetura que está sendo utilizada. 
O processador efetua uma computação por meio do cilo de busca e de execução. 

Cada instrução instrui o processador a realizar alguma operação. Por exemplo: some 2 números, compare 2 números, etc.

1) O processador lê uma instrução da memória;
2) Executa a instrução e lê uma próxima;




## X86 - Arquitetura da Intel

CPU:
+ Registradores
    + Eles guardam os valores;
    + Como uma variável;
    + Uma coleçã de circuitos integrados (CIs) que são capazes de armazenar bits;
        + A capacidade de um registrador é um atributo importante do processador;
        + Se o processador não consegue armazenar a quantidade necessária para realizar a operação de forma completa, a operação é quebrada;
+ ULA - Unidade Lógica e Artimética
    + Realiza todas as operações lógicas (and, or, xor, ...);
    + Operações aritméticas (adição, subtração...);
+ UC - Unidade de Controle
    + Ela controla as outras duas unidades;

Registradores - Existem vários tipos:
1) Contador de Programa (PC) - Ele aponta para a próxima instrução;
2) Registrador de Instrução (IR) - Ele guarda a instrução que o processador acabou de carregar;
3) Registradores de Propósito Geral - Armazenamento de resultados intermediários;

As instruções podem ser de vários tipos:
+ Operações aritméticas e lógicas;
    + Na maioria das vezes elas são aplicadas nos registradores;
+ Movimentações de dados;
+ Operações de Entrada e Saída;
+ Controle do fluxo de execução (desvios condicionais e incondicionais);
