### Cálcula um hash fácil (Python):

```python
# Escreva um programa em Python que recebe uma string de 1 a 90
# caracteres ASCII da entrada padrão e calcula o hash dessa string
# seguindo estes passos:
#
# +++ Passo 1: Ajuste de tamanho +++
#          Esse passo faz com que a string tenha um tamanho que seja
#          um múltiplo de 3. Para isso, calcula quantas posições devem
#          ser preenchidas assim: 3-tamanho(string)%3 . Mesmo se a
#          string já tiver um tamanho que seja múltiplo de 3, vai
#          preencher mais 3 posições com o valor 3 (obs.: de fato é o
#          valor 3 em decimal e não o caracter ASCII '3')
#
# +++ Passo 2: Cálculo e concatenação dos XOR +++
#          Esse passo aumenta a string do passo anterior colocando
#          mais 3 caracteres nela. Será copiado o maior caracter da
#          string 3 vezes
#
# +++ Passo 3: Transformação dos blocos em apenas 1 bloco +++
#          Esse passo considera a string como uma matriz de 3 colunas
#          e faz o XOR de cada coluna gerando um vetor final de 3
#          posições que é justamente o hash buscado
#
# +++ Passo 4: Definição do hash como um valor em hexadecimal +++
#          Imprime na tela a saída do passo anterior mas como
#          hexadecimal. Lembrando que cada valor do vetor tem que
#          ocupar exatamente 2 caracteres. Portanto, um valor menor
#          que 16 (10 em hexadecimal) precisa ter um zero mais à
#          esquerda
#
# Obs.: as impressões intermediárias estão no código para fins
# didáticos e para facilitar a depuração          

def main():
    # Passo 0: Lendo a string da entrada padrão
    entrada=input("")
    
    # Passo 1: Ajustando o tamanho da mensagem
    tamanho=len(entrada)
    quantofalta=3-tamanho%3
    
    entradaEmAscii=[0]*tamanho
    for i in range(tamanho):
        entradaEmAscii[i]=ord(entrada[i])
    saidaPasso1 = entradaEmAscii + quantofalta*[3]

    print("Saida do Passo 1:")
    print(saidaPasso1)
     
    # Passo 2: encontrando o maior código ASCII e contatenando ele
    # 3 vezes à saidaPasso1
    
    saidaPasso2 = saidaPasso1 + 3*[max(saidaPasso1)]

    print("Saida do Passo 2:")
    print(saidaPasso2)
 
    # Passo 3: faz o XOR dos valores (seriam os valores da mesma
    # coluna se o array estivesse sendo visto como uma matriz) e esse
    # é o hash
    saidaPasso3 = saidaPasso2[:3]

    n = len(saidaPasso2)//3
    for j in range(3):
        for i in range(1,n):
            saidaPasso3[j] = saidaPasso2[i*3+j] ^ saidaPasso3[j]

    print("Saida do Passo 3:")
    print(saidaPasso3)

    # Passo 4: imprime o vetor do Passo 3 em hexadecimal
    for i in range(3):
        if saidaPasso3[i]<16:
            print("0"+hex(saidaPasso3[i])[2:],end="")
        else:
            print(hex(saidaPasso3[i])[2:],end="")
    print()

main()
```
